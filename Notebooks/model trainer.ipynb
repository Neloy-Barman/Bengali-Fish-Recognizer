## **Fetching dataloader**

```python
data_loader = torch.load(f"{data_loader_path}/DataLoader_v_{version}")
```

```python
data_loader.train.show_batch(max_n=20, nrows = 4)
```

```python
data_loader.valid.show_batch(max_n=20, nrows = 4)
```

```python
model = vision_learner(data_loader, vgg19_bn, matrics = [error_rate, accuracy])
```

```python
model.fine_tune(10)
```

```python
model_name = "VGG19"
model_version = 1
```

```python
model.save(f"{model_name}_{model_version}")
```
